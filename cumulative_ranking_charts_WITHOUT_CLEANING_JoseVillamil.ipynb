```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import auth
# get credentials
auth.authenticate_user()

query = '''
SELECT
  item_category,
  COUNT(item_category) AS count_item_category,
  RANK() OVER (ORDER BY COUNT(item_category) DESC) AS rank_item_category
FROM
  `operation-enchante-fans.franquicias_pro_fans.order_line`
GROUP BY item_category
ORDER BY count_item_category DESC, item_category;
'''
# order_line ranking por categor√≠a de item
ol_rank_itemcat = pd.read_gbq(query, project_id="operation-enchante-fans")
ol_rank_itemcat.head()

```

```python
import plotly.express as px

fig = px.bar(ol_rank_itemcat, x='item_category', y='count_item_category')
fig.show()
```
