````python
order_data_df.groupby(["id_store"])["price"].sum().sort_values(ascending=False)
````



```
 average price of an order for each restaurant
```



````python
df_turnover = order_data_df.groupby(["id_store"])["price"].agg(['size', 'sum']).reset_index()
df_turnover["avg_order_price"] = df_turnover["sum"] / df_turnover["size"]
df_turnover.sort_values("avg_order_price", ascending=False)
````
